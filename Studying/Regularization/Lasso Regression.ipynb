```python
import pandas as pd

dataset = pd.read_csv("data/ridge.csv")
dataset_copy = dataset.copy()

X = dataset.iloc[:, :-1]
Y = dataset.iloc[:, -1]
```

```python
X_plt = [0, 1, 2, 3, 4]
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso

alfas = [0, 0.5, 1, 2, 4]

for i, l in enumerate(alfas):
    lasso = Lasso(alpha = l)
    lasso.fit(X, Y)
    
    Y_plt = lasso.predict(pd.DataFrame(X_plt))
    
    plt.figure(2)
    plt.plot(X_plt, Y_plt)
    plt.ylim(ymin=0, ymax=9)
    plt.xlim(xmin=0, xmax=6)

    plt.text(X_plt[-1], Y_plt[-1], 'y = '+ str('%.2f'%lasso.coef_) + '* x + ' + str('%.2f'%lasso.intercept_) + ' for \u03BB = ' + str(l))
```
