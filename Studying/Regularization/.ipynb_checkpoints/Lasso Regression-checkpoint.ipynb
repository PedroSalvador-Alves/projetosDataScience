```python
import pandas as pd

dataset = pd.read_csv("data/ridge.csv")
dataset_copy = dataset.copy()

X = dataset.iloc[:, :-1]
Y = dataset.iloc[:, -1]
```

```python
from sklearn.linear_model import Ridge
alfa = 0
#alfa = 0, é a regressão linear comum
ridge = Ridge(alpha = alfa)

ridge.fit(X, Y)
```

```python
ridge.coef_
```

```python
ridge.intercept_
```

```python
X_plt = [0, 1, 2, 3, 4]
Y_plt = ridge.predict(pd.DataFrame(X_plt))
```

```python
import matplotlib.pyplot as plt

plt.figure(1)
plt.plot(X_plt, Y_plt)
plt.ylim(ymin=0, ymax=9)
plt.xlim(xmin=0, xmax=6)

plt.text(X_plt[-1], Y_plt[-1], 'y = '+ str('%.2f'%ridge.coef_) + '* x + ' + str('%.2f'%ridge.intercept_) + ' for \u03BB = ' + str(alfa))
```

```python
alfas = [0, 1, 10, 100]

for i, l in enumerate(alfas):
    ridge = Ridge(alpha = l)
    ridge.fit(X, Y)
    
    Y_plt = ridge.predict(pd.DataFrame(X_plt))
    
    plt.figure(2)
    plt.plot(X_plt, Y_plt)
    plt.ylim(ymin=0, ymax=9)
    plt.xlim(xmin=0, xmax=6)

    plt.text(X_plt[-1], Y_plt[-1], 'y = '+ str('%.2f'%ridge.coef_) + '* x + ' + str('%.2f'%ridge.intercept_) + ' for \u03BB = ' + str(l))
```
